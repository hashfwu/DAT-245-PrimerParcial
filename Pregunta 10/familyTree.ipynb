# Usando funciones y solo la relación padre/madre

*Creando las funciones para la definición de relaciones basadas en padre/madre*

```python
pa_madre_F=Relation()
facts(pa_madre_F, ("Justo", "Wilfredo"), ("Justo", "Martha")
,("Isabel_V", "Wilfredo"), ("Isabel_V", "Martha")
,("Dionicio", "Victoria")
,("Victoria_A", "Victoria")
,("Hugo", "Eliana"), ("Hugo", "Victor")
,("Martha", "Eliana"), ("Martha", "Victor")
,("Wilfredo", "Rodrigo"), ("Wilfredo", "Isabel")
,("Victoria", "Rodrigo"), ("Victoria", "Isabel"))
```

```python
def abuelo_a_F(x, z):
    y = var()
    return lall(pa_madre_F(x, y), pa_madre_F(y, z))
```

###  🤖Código:
Retorna una relacion entre **x** y **z** tal que:
- La relación entre **x** y **y** sea de padre/madre y la relación de **y** y **z** sea igual de padre/madre, donde **y** es una variable lógica

>The examples above made implicit use of the goal constructors lall and lany, which represent goal conjunction and disjunction, respectively. <p>
>*Los ejemplos anteriores hicieron un uso implícito de los constructores de objetivos lall y lany, que representan la conjunción y disyunción de objetivos, respectivamente.* <p>
[Ropositorio de Kanren en GitHub](https://github.com/pythological/kanren)

```python
print(run(0, x, abuelo_a_F(x, 'Isabel')))
print(run(0, x, abuelo_a_F(x, 'Rodrigo')))
print(run(0, x, abuelo_a_F(x, 'Eliana')))
```

En efecto, como ambos somo hermanos compartimos los mismo abuelos. Veamos la relación de hermanos:

```python
def hermano_a_F(x,z):
    y=var()
    return lall(pa_madre_F(y, x), pa_madre_F(y, z))
```

```python
print(run(0, x, hermano_a_F(x, 'Rodrigo')))
print(run(0, x, hermano_a_F('Wilfredo', x)))
print(run(0, x, hermano_a_F(x, 'Eliana')))
```

###  🤖Código:
Se podria pensar que este código es redundante por que devuelve el mismo valor **dos veces**. **No es asi**, lo que pasa es que en este caso, cada individuo tiene **2 padres** y por lo tanto la relación de hermanos se repide **dos veces**. Para arreglar eso podriamos límitar el código nuevemente para que solo se limite a una relación:

```python
print(run(1, x, hermano_a_F(x, 'Rodrigo')))
print(run(1, x, hermano_a_F('Wilfredo', x)))
print(run(1, x, hermano_a_F(x, 'Eliana')))
```

Pero vemos que en este caso 'Eliana' es la hermana de 'Eliana' por que el padre de 'Eliana' es tambíen el padre de 'Eliana'. Para solucionar esto, reparamos la función:

```python
def hermano_a_F(x,z):
    y=var()
    return lall(pa_madre_F(y,x), pa_madre_F(y,z), neq(x,z))
    # neq hace que x y z no sean iguales
print(run(0, x, hermano_a_F(x, 'Rodrigo')))
print(run(0, x, hermano_a_F('Wilfredo', x)))
print(run(0, x, hermano_a_F(x, 'Eliana')))
```

Retorna una relación tal que el padre de **x** sea **y** y el padre de **z** sea igual **y**. Claro que retorna dos resultados, por que la relación con un hermano puede ser por padre o por madre. Con esto podemos crear la relaición de tios:

```python
def tio_a_F(x, z):
    y=var()
    return lall(hermano_a_F(x, y), pa_madre_F(y,z))
```

Retorna la relacion entre **x** y **z** tal que **x** sea hermano de la variabla lógica **y** y que **y** sea el padre de **z**. A continuación veremos que se puede producir:

```python
print('Los tios de Eliana son: ')
print(run(0, x, tio_a_F(x, 'Eliana')))
print('Los tios de Isabel')
print(run(0, x, tio_a_F(x, 'Isabel')))
print('Los tios de Martha son: ')
print(run(0, x, tio_a_F(x, 'Martha')))
print('Esto es por que aun no hemos definido los tios de Martha')
```
